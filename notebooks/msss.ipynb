```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

plt.rcParams['image.origin'] = 'upper'
```

```python
# Load the image
folder = "/Users/mcha5804/Library/CloudStorage/OneDrive-TheUniversityofSydney(Students)/Resources/Prezzos/MSSS2023/"
hubble_path = folder + "hubble.png"
img = Image.open(hubble_path)

# Convert to grayscale
gray_img = img.convert('L')

# Convert to numpy array
eagle_neb = np.array(gray_img)
eagle_neb = eagle_neb / eagle_neb.sum()
```

```python
# Plot with viridis colormap
plt.imshow(eagle_neb, cmap='plasma')
plt.axis('off')
plt.colorbar(label="Relative Intensity")
# plt.savefig('hubble_plasma.pdf', dpi=1000, bbox_inches='tight')
plt.show()
```

```python
creeper_path = folder + "creeper.png"
img = Image.open(creeper_path)

# Convert to grayscale
gray_img = img.convert('L')

# Convert to numpy array
creeper = np.array(gray_img, dtype=int)

fig, ax = plt.subplots()
mat = ax.imshow(creeper, vmax=1e10, cmap="Greys")  # Display matrix
for (i, j), z in np.ndenumerate(creeper):  # Loop over matrix
    ax.text(j, i, '{:.0f}'.format(z), ha='center', va='center')
ax.axis('off')
plt.savefig('creeper_matrix.pdf', dpi=1000, bbox_inches='tight', transparent=True)
plt.show()
```
